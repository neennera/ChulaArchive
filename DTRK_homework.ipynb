## HW1 : เขียนเกมเป้ายิ้งฉุยด้วย Python

```python
# set up
import random
player_score = {
    'win' : 0,
    'lose' :0,
    'draw' : 0
}

# rock << paper << scissors << rock
# 1 << 2 << 3 << 1
choice_map = {}
choice_map[1] = 'rock'
choice_map[2] = 'paper'
choice_map[3] = 'scissors'

choice_map['rock'] = 1
choice_map['paper'] = 2
choice_map['scissors'] = 3

def play() :
    while(1) :
        player_input = input("pick rock/paper/scissors")
        if (player_input in choice_map) and (player_input>"3"):
            player_choice_int = choice_map[player_input]
            player_choice = player_input
            break
        else :
            print("try again")
            continue
    machine_choice_int = random.randint(1,3)
    machine_choice = choice_map[machine_choice_int]
    print("You picked :",player_choice)
    print("I picked :", machine_choice)

    if (player_choice_int == machine_choice_int) :
        print("Draw")
        player_score['draw'] +=1
    elif (player_choice_int +1 == machine_choice_int) or (player_choice_int==3 and machine_choice_int==1) :
        print("you lose")
        player_score['lose'] +=1
    else :
        print("you win")
        player_score['win'] +=1

def games() :
    name = input("input your name")
    while(input("Wanna play? (y/n)") == 'y') :
        play()
    for k in player_score :
        print(name, "had", k, player_score[k], "time"+ ("s" if player_score[k]>1 else ""))
```

```python
# run this cell to play
games()
```

## HW2 : เขียน class ATM มีอย่างน้อย 5 methods

```python
import random
class ATM :
    def __init__(self, name, bank) :
        self.name = name
        self.bank = bank
        self.balance = 0
    
    def Deposit(self,money) :
        if (money<0) :
            print("!!!! money should be more than 0")
            return
        self.balance += money
        print(">>>> Your balance is", self.balance)

    def Withdraw(self,money) :
        if (money<0) :
            print("!!!! money should be more than 0")
            return
        if(money > self.balance) :
            print("!!!! not enough money")
        else :
            self.balance -= money
            print(">>>> Your balance is", self.balance)
    
    def transfer(self,money, targetBankNumber) :
        if (money<0) :
            print("!!!! money should be more than 0")
            return
        if(money > self.balance) :
            print("!!!! not enough money")
        else :
            self.balance -= money
            print(">>>> You had transfer", money, "bath to", targetBankNumber)
    
    def Interest(self) :
        print(">>>> You has a yearly interest", 0.05*self.balance)
        self.balance = 1.05*self.balance
        print(">>>> Your balance is", self.balance)
    
    def OTP(self) :
        otp = random.random()
        otp = int((otp*10**6) // 1)
        print(">>>> your OTP is", otp, "(dont share!)")
```

```python
name = input("type your name")
bank = input("which bank you want to open account with?")
b1 = ATM(name, bank)

while(1) :
    print("choose the services (1-6)")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Transfer")
    print("4. Interest")
    print("5. OTP")
    print("6. exit")
    op = int(input())

    try :
        if(op==1) :
            money = int(input("insert the money : "))
            b1.Deposit(money)
        elif(op==2) :
            money = int(input("insert the money : "))
            b1.Withdraw(money)
        elif(op==3) :
            money = int(input("insert the money : "))
            bank = input("insert the target bank number : ")
            b1.transfer(money,bank)
        elif(op==4) :
            b1.Interest()
        elif(op==5) :
            b1.OTP()
        elif(op==6) :
            break
        else :
            print(">>>> try again")
    except :
        print(">>>> try again")
```

## HW3 :


```python
import requests
from gazpacho import Soup
import time
import pandas as pd

df = pd.DataFrame()


url = f"https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"
res = requests.get(url)

html = Soup(res.text)
```

```python
title = html.find("h3", {"class":"lister-item-header"})
title = [x.strip() for x in title]

rating = html.find("div", {"class":"ratings-imdb-rating"})
rating  = [float(x.strip()) for x in rating]

year = [x[x.find("(")+1 : x.find(")")] for x in title]
for i in range(len(year)) :
    if(len(year[i]) < 4) :
        st, ed = 0,0
        while(1) :
            st = title[i].find("(", st+1)
            ed = title[i].find(")", ed+1)
            if(st==-1 or ed==-1) :
                break
            year[i] = title[i][st+1 : ed]

clean_title = [x.split("(")[0] for x in title]
clean_title = [x.split(". ")[1] for x in clean_title]
```

```python
df = pd.DataFrame( data = 
    {
        "title" : clean_title,
        "rating" : rating,
        "launched year" : year
    }
)
df
```

## HW4 : เขียนโค้ดดึง public API + ใช้ requests และ time.sleep

```python
import requests
import time
import pandas as pd

df = pd.DataFrame()

for i in range(15) :
    url = f"https://pokeapi.co/api/v2/pokemon/{i+1}"

    res = requests.get(url)
    if(res.status_code == 200) :
        data = res.json() 
        dff = pd.DataFrame.from_records(
            [
            {
            "id" : i+1,
            "name" : data['name'],
            "weight" : data['weight'], 
            "height" : data['height'],
            "abilities" : data['abilities'][0]['ability']['name'], 
            }
            ], index = "id"
        )
        df = pd.concat([df, dff])
        time.sleep(1)
```
